```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# Generiere einen knstlichen Datensatz
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Konvertiere X und y in einen DataFrame
data = pd.DataFrame(X)
data['target'] = y

# Zeige die ersten Zeilen des DataFrame
print(data.head())

# Teile den Datensatz in Trainings- und Testdaten
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Listen fr die Speicherung der Genauigkeiten
train_accuracy = []
test_accuracy = []

# Trainiere Decision Tree Classifier mit unterschiedlicher Tiefe
for depth in range(1, 15):
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    train_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

# Plotte die Genauigkeiten
plt.figure(figsize=(12, 6))
plt.plot(range(1, 15), train_accuracy, label='Trainingsgenauigkeit')
plt.plot(range(1, 15), test_accuracy, label='Testgenauigkeit')
plt.xlabel('Baumtiefe')
plt.ylabel('Genauigkeit')
plt.title('Trainings- und Testgenauigkeit bei unterschiedlichen Baumtiefen')
plt.legend()
plt.show()

# PCA zur Reduzierung auf zwei Dimensionen
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotte den Datensatz nach PCA
plt.figure(figsize=(12, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=40)
plt.title('2D-Visualisierung des Datensatzes nach PCA')
plt.xlabel('Hauptkomponente 1')
plt.ylabel('Hauptkomponente 2')
plt.colorbar(label='Klasse')
plt.show()
```
